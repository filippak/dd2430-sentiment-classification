```python
import matplotlib.pyplot as plt
import numpy as np
import collections
import pprint
```

```python
training_results = np.load('results_5_percent.npy', allow_pickle=True)
batches  = list(range(1, 6))
generator_loss = []
disc_loss = []
for idx, obj in enumerate(training_results[0]):
    # Append the loss for the first run of the experiment
    generator_loss.append(obj['Training Loss generator'])
    disc_loss.append(obj['Training Loss discriminator'])
```

```python
# Plot generator and discriminator loss 
plt.figure(figsize=(8,6))
plt.xticks(batches)
plt.plot(batches, generator_loss, label='Generator')
plt.plot(batches, disc_loss, label='Discriminator')
plt.xlabel('epoch')
# Set the y axis label of the current axis.
plt.ylabel('average loss')
# Set a title of the current axes.
plt.title('Loss for generator and discriminator with 5% labeled data')
plt.legend()
plt.show()
plt.savefig("loss_plot")
```
